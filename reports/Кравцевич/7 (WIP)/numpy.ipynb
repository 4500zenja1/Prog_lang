```python
import math
import numpy as np
import imageio
from scipy.spatial.distance import euclidean
import matplotlib.pyplot as plt
```

```python
x = np.array(
		[[1, 0, 1],
		 [2, 0, 2],
		 [3, 0, 3],
		 [4, 4, 4]])
%timeit x.diagonal()[x.diagonal() != 0].prod()
```

```python
x = np.array([[9, 4, 2], [6, 0, 0], [9, 9, 3]])
i = np.array([1, 2, 1])
j = np.array([1, 0, 1])

%timeit x[i, j]
```

```python
x = np.array([1, 2, 2, 4])
y = np.array([4, 2, 1, 2])
x.sort()
y.sort()
%timeit all(x == y)
```

```python
coefs = [0.299, 0.587, 0.114]
img = imageio.imread("1.png")
plt.imshow(img)
```

```python
reshaped_image = img[:, :, :-1]
%timeit new_img = (reshaped_image * (([[coefs] * reshaped_image.shape[1]]) * reshaped_image.shape[0])).sum(axis=2)
```

```python
plt.imshow(new_img)
```

```python
x = np.array([2, 2, 2, 3, 3, 5])
result = np.unique(np.array(x), return_counts=True)
%timeit np.unique(np.array(x), return_counts=True)

print(*result, sep='\n')
```

```python
x = np.array([2, 7, 6, 6, 9, 6, 3, 4, 9])
y = np.array([1, 0, 0, 7, 2, 2, 4, 3, 0])
%timeit math.sqrt(((x - y) ** 2).sum())
%timeit np.linalg.norm(x - y)
%timeit euclidean(x, y)
```
